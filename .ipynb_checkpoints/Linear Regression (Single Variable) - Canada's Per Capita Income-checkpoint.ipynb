# Machine Learning Practice

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data=pd.read_csv("data.csv")
```

```python
data
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("income (USD)")
plt.scatter(data.year,data.income,color="red",marker="+")
```

```python
reg=linear_model.LinearRegression()
reg.fit(data[["year"]],data.income)
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("income (USD)")
plt.scatter(data.year,data.income,color="red",marker="+")
plt.plot(data.year,reg.predict(data[["year"]]),color="blue")
```

```python
years=pd.read_csv("years.csv")
```

```python
years
```

```python
income=reg.predict(years)
```

```python
income
```

```python
years["income"]=income
```

```python
years
```

```python
years.to_csv("result.csv",index=False)
```
